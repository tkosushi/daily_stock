```python
# 
import pandas as pd
import matplotlib.pyplot as plt

# CSV
data = pd.read_csv('data/7974.T_20230213-20230411.csv')
data['Date'] = pd.to_datetime(data['Date'])  # Datedatetime

# 
print(data.head())

# 
fig, ax1 = plt.subplots(figsize=(14, 7))  # 

# 
ax1.plot(data['Date'], data['Close'], color='b')  # 
ax1.set_title('Cup and Handle Pattern')  # 
ax1.set_xlabel('Date')  # x
ax1.set_ylabel('Close Price', color='b')  # y
ax1.grid(True)  # 

# 
ax2 = ax1.twinx()  # 2Y
ax2.bar(data['Date'], data['Volume'], color='gray', alpha=0.3)  # 
ax2.set_ylabel('Volume', color='gray')  # y

plt.show()  # 
```
